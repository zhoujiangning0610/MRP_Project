```python
import pandas as pd
from sklearn import linear_model    
df = pd.read_excel("Clean_Data.xlsx")
df.index = df["DateTime"]
df.index = pd.to_datetime(df.index)
df
```

```python
df = df.replace("heatOff", 0)
df = df.replace("heatStage1On", 1)
df = df.replace("Home", 1)
df = df.replace("Sleep", 0)
need_df = df[["System Mode","Program Mode","Heat Set Temp (C)","Current Temp (C)","Current Humidity (%RH)","Outdoor Temp (C)","Thermostat Temperature (C)","Thermostat Humidity (%RH)","Bedroom (C)","3rd floor landing (C)","3rd floor room (C)","Basement (C)","Heat Stage 1 (sec)"]]
need_df
```

```python
df_hour = need_df.resample(rule = 'H').sum()
df_hour.isnull().sum()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
model = linear_model.LinearRegression()
X = df_hour.iloc[:,:-1]
y = np.array(df_hour.iloc[:,-1])
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)

sc = StandardScaler()
sc.fit(y.reshape(-1,1))
y = sc.transform(y.reshape(-1,1))

model.fit(X,y)
predict_y = model.predict(X)

rmse = np.sqrt(mean_squared_error(y,predict_y))

mae = mean_absolute_error(y, predict_y)
print(rmse,mae)
```
